Step 1

Null Hypothesis is want we want to prove

Step 2

alpha = 0.05
As it is a two-tailed test

```python
import statsmodels.api as sm
```

```python
sm.qqplot(cutlets['Unit A'],line = 'q')
plt.title('Unit A')
sm.qqplot(cutlets['Unit B'],line = 'q')
plt.title('Unit B')
plt.show()
```

compare evidences with t statistics

```python
statistic,p_value = stats.ttest_ind(cutlets['Unit A'],cutlets['Unit B'],alternative = 'two-sided')
print('p_value=',p_value)
```

Compare p_value with 'α '(Significane Level)
If p_value is ≠ 'α ' we failed to reject Null Hypothesis because of lack of evidence
If p_value is = 'α' we reject Null Hypothesis
interpreting p-value


```python
alpha = 0.025
print('significance=%.3f,p=%.3f' % (alpha,p_value))
if p_value <= alpha:
    print('we reject null hypothesis there is significane difference btw two units')
else:
    print('we fail to reject null hypothesis')
```

Question 3
Hypothesis Testing Exercise!

      Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.
      
We are going to conduct a Test of Independence using Chi-Square ᵪ2 test with Contingency table

We need to check whether the proportion of any of these samples are different or the same?

Step 1

Make two Hypothesis one contradicting to other
Null Hypothesis is want we want to prove
Null Hypothesis: There is no association or dependency between the gender based buyer rations across regions
Alternative Hypthosis: There is a significant association or dependency between the gender based buyer rations across regions

Step 2
Decide a cut-off value
Significance 5%
alpha = 0.05
As it is a one-tailed test
alpha = 1-0.95 = 0.05

Step 3
Collect evidence

Importing Files
 

```python
buyer = pd.read_csv('BuyerRatio.csv')
buyer
```

```python
table = [[50,142,131,70],
         [435,1523,1356,750]]
```

applying chi square contigece table to convert observed value into expected value

```python
buyer1=buyer.drop(columns='Observed Values',axis=True)

stat, p, dof, exp = stats.chi2_contingency(buyer1)
print(stat,"\n",p,"\n",dof,"\n",exp)
```

```python
stats.chi2_contingency(table)
```

```python
observed = np.array([50,142,131,70,435,1523,1356,750])
expected = np.array([42.76531299,146.81287862,131.11756787,72.30424052,442.23468701,1518.18712138,1355.88243213,747.69575948])
```

comparing evidence with hypothesis

```python
statistics,p_value = stats.chisquare(observed,expected,ddof = 3)
print("Statistics = ",statistics,"\n",'P_Value = ', p_value)
```

Compare p_value with 'α '(Significane Level)

If p_value is ≠ 'α ' we failed to reject Null Hypothesis because of lack of evidence

If p_value is = 'α' we reject Null Hypothesis

interpreting p-value


```python
alpha = 0.05
print('significance=%.3f , p=%.3f'%(alpha,p_value))
if p_value <= alpha:
    print('we reject null hypothesis there is a significant difference')
else:
    print('we fail to reject null hypothesis')
```
